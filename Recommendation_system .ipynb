```python
import ast 
```

```python
def convert(text):
    synt=[]
    for i in ast.literal_eval(text):
        synt.append(i['name'])
    return synt
    
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'][0]
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert5(text):
    syntex=[]
    counter=0
    for i in ast.literal_eval(text):
        if counter < 5:
            syntex.append(i['name'])
        counter += 1
    return syntex
```

```python
movies['cast']=movies['cast'].apply(convert5)
```

```python
movies['crew'][0]
```

```python
def director(text):
    dir=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            dir.append(i['name'])
    return dir 
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies['crew'][0]
```

```python
movies.head()
```

```python
def collapse(A):
    A1=[]
    for i in A:
        A1.append(i.replace(' ',''))
    return A1
```

```python
movies.columns
```

```python
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
new=movies.drop(columns=['genres',  'keywords',  'overview', 'cast', 'crew'])
```

```python
# now we have cleaned data for buliding recommendation sysytem
```

```python
# now feature engineering required to convert non numeric data to numeric data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features = 5000 ,stop_words='english',binary=True)
vector=cv.fit_transform(new['tages']).toarray()
```

```python
vector.shape
```

```python
vector
```

```python
# building model-- cosin similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
similarity.shape
```

```python
# building recommendation system 
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:11]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Spider-Man')
```

```python
recommend('Kung Fu Panda 3')
```
